```python
events.columns = ['date','device_id','user_id','server_id','connection_state']
```

```python
events['date'] = pandas.to_datetime(events['date'], errors='coerce')
```

```python
index = 0
# get dates column
all_days = events['date'].dt.day
# first day value
first_day = all_days[0]
# iterate through data as long as day is the same
for day in all_days:
    if day != first_day:
        break
    index += 1
# one day events list
one_day_events = events[0:index]

# # print
# one_day_events
    
    
```

```python
one_day_devices = one_day_events['device_id']
```

```python
only_first_device_connection = one_day_devices.drop_duplicates()
```

```python
user_dic = {}

for event_index in range(len(only_first_device_connection)):
    # get real index of first device connection
    selected_index = only_first_device_connection.index[event_index]
    # print('index: {},   user: {}'.format(selected_index, events.iloc[selected_index]['user_id']))
   
    user_index = events.iloc[selected_index]['user_id']
    # if above event is in built dictionary
    if user_index in user_dic:
        # increment usres fleet
        user_dic[user_index] += 1
    # if no
    else:
        # add user to dic
        user_dic[user_index] = 1
        
        
        
user_dic
        
    
    
```
